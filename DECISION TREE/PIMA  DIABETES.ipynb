```python
import pandas as pd
```

```python
pima=pd.read_csv(r"/users/pawarbharath/downloads/decision/diabetes.csv")
```

```python
pima
```

```python
x=pima.iloc[:,0:8]
y=pima.diabetes_class
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix


print(classification_report(y_test,y_pred))
```

```python
clf1 = DecisionTreeClassifier(criterion = "entropy", splitter = "random", max_depth = 20, min_samples_split = 10,min_samples_leaf = 2, max_features = 5)
```

```python
clf1.fit(x_train,y_train)
```

```python
y_pred=clf1.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

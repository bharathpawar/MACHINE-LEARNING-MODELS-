```python
import pandas as pd
```

```python
boston = pd.DataFrame(boston_dataset.data,columns=boston_dataset.feature_names)
```

```python
boston['MEDV'] = boston_dataset.target
```

```python
corelmatrix=boston.corr().round(2)
```

```python
import seaborn as sns
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
sns.heatmap(data=corelmatrix,annot=True,cmap="coolwarm")
```

```python
sns.scatterplot(boston.LSTAT,boston.MEDV)
```

```python
sns.scatterplot(boston.RM,boston.MEDV)
```

```python
x=boston[['LSTAT','RM']]
y=boston[['MEDV']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lrm1=LinearRegression()
```

```python
lrm1.fit(x_train,y_train)
```

```python
lrm1.predict(x_test)
```

```python
lrm1.score(x_test,y_test)
```
